```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the gradient descent algorithm and necessary function
def objective(x):
    return x**2

def derivative(x):
    return 2*x

def gradient_descent(n_iter, initial_value, learning_rate):
    x_list = []
    y_list = []
    x = initial_value
    
    for i in range(n_iter):
        slope = derivative(x)
        x = x - learning_rate*slope
        y = objective(x)
        x_list.append(x)
        y_list.append(y)
        print(f'iter:{i+1}| f({x}) = {y:.10f}')
        
    print('Iterasi selesai!')
    print(f'f({x}) = {y}')
    
    return x_list, y_list

best, score = gradient_descent(100, 5, 0.1)
```

```python
# plot the gradient descent iteration
x_axis = np.linspace(-5, 5, 100)
plt.plot(x_axis, objective(x_axis), label='objective function: $x^2$')
plt.plot(best, score, '.-', label='gradient descent plot', color='red')
plt.axhline(y=np.min(score), label='global minimum', color='gray', linestyle='--')
plt.legend()
plt.show()
```
