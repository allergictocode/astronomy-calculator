# **Rayleigh-Jeans Law**

### Domain panjang gelombang

$$B_λ(T)=\frac{2ck_BT}{λ^4}$$

### Domain frekuensi

$$B_ν(T)=\frac{2\nu^2k_BT}{c^2}$$


# **Planck's Law**

### Domain panjang gelombang

$$B_λ(λ, T)=\frac{2hc^2}{λ^5}\frac{1}{e^{\frac{hc}{\lambda k_BT}}-1}$$

### Domain frekuensi

$$B_ν(ν, T)=\frac{2hν^3}{c^2}\frac{1}{e^{\frac{hν}{k_BT}}-1}$$


# **Wien's Distribution Law**

### Domain panjang gelombang

$$I(λ, T)=\frac{2hc^2}{λ^5}e^{-\frac{hc}{\lambda k_BT}}$$

### Domain frekuensi

$$I(ν, T)=\frac{2hν^3}{c^2}e^{-\frac{hν}{k_BT}}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Variables
c = 299792458           # m/s
k = 1.380649*10**-23    # J/K
h = 6.62607015*10**-34  # J/Hz
```

#Domain panjang gelombang

```python
# Rayleigh-Jeans Law
def RJ_l(l, T):
  B = (2*c*k*T) / (l**4)
  return B

# Planck's Law
def P_l(l, T):
  B = (2*h*(c**2) / l**5) * (1/(np.exp(h*c/(l*k*T))-1))
  return B

# Wien's distribution Law
def W_l(l, T):
  I = (2*h*(c**2) / l**5) * np.exp(-(h*c)/(l*k*T))
  return I
```

##6000 K

```python
# Plot the graph
x = np.linspace(0, 3*10**-6, 1000)
y_RJ = RJ_l(x, 6000)
y_P = P_l(x, 6000)
y_W = W_l(x, 6000)
plt.plot(x, y_RJ, label='Rayleigh-Jeans', linestyle=':')
plt.plot(x, y_P, label='Planck', linestyle='-')
plt.plot(x, y_W, label='Wien', linestyle='--')
plt.ylim(0, 4*10**13)
plt.xlabel('Wavelength (m)')
plt.ylabel('Intensity')
plt.title('T = 6000 K')
plt.grid()
plt.legend()
plt.show()
```

#Domain frekuensi

```python
# Rayleigh-Jeans
def RJ_f(f, T):
  B = 2*(f**2)*k*T / c**2
  return B

# Planck
def P_f(f, T):
  B = (2*h*(f**3)/c**2) * (1/(np.exp((h*f)/(k*T))-1))
  return B

# Wien
def W_f(f, T):
  I = 2*h*(f**3)/c**2 * np.exp(-(h*f)/(k*T))
  return I
```

## T = 6000 K

```python
# Plot the graph
x = np.linspace(0, 5*10**15, 1000)
y_RJ = RJ_f(x, 6000)
y_P = P_f(x, 6000)
y_W = W_f(x, 6000)
plt.plot(x, y_RJ, label='Rayleigh-Jeans', linestyle=':')
plt.plot(x, y_P, label='Planck', linestyle='-')
plt.plot(x, y_W, label='Wien', linestyle='--')
plt.ylim(0, 6*10**-8)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Intensity')
plt.title('T = 6000 K')
plt.grid()
plt.legend()
plt.show()
```
